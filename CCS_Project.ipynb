```python
from math import pow
def calc_probability(current_city, cities, pheromones, visited_cities):
  prob = np.zeros_like(cities)
  z=0
  for i in range(cities.shape([0])):
    if i not in visited_cities:
      z += (pheromones[current_city][i])*(1/(cities[current_city][i]))
  for i in range(cities.shape([0])):
    prob[current_city][i] = (pow(pheromones[current_city][i],1))*(pow(1/(cities[current_city][i]),2)) / (z)
  return prob
```

```python
def move(current_city, cities, pheromones, visited_cities):
  min_cost = 0
  max_prob = 0
  next_city
  for i in range(cities.shape([0])):
    prob = calc_probability(current_city, cities, pheromones, visited_cities)
    if prob[current_city][i] > max_prob:
      max_prob = prob[current_city][i]
      next_city = i
    #calc th ecost using the formula
    path_cost = cities[current_city][i]
    if min_path == 0:
      min_path = path_cost
    else:
      if path_cost < min_path:
        min_path = path_cost
    return next_city, min_cost
```

```python
def start_end(cities, n_ant, pheromones):
  n_cities = len(cities)
  # start_pos = (0,0)
  visited_cities=[0]
  current_city= 0
  total_cost = 0
  while len(visited_cities) < n_cities:
    next_city, cost = move(current_city, cities, pheromones, visited_cities)
    total_cost += cost
    visited_cities.append(next_city)
    current_city = next_city
# must return to the starting point

  return visited_cities, total_cost
```
